```python
def normal_dist(n): # n: number of random numbers generated
    '''
    Outputs list of randomly generated numbers from a normal distribution.
    n: number of random numbers generated or length of output list
    '''
    lists = [] 
    for i in range(1, 12*n, n): 
        x = lists.append(list([rnd.random() for i in range(n)]))
        out = sum(map(np.array,lists)) / 12 # divide by 12 to normalize
    return out
    
    
```

## Problem 3

```python
def ran_vector(dim):
    '''
    Takes in a dimension and returns randomly generated unit vector of
    that dimension from uniform distribution with mean 0.
    
    dim: number of dimensions for generated vectors
    
    returns:
        a numpy array containing a randomly generated unit vector
    '''
    vec = []
    for i in range(dim):
            uni_x = rnd.random()
            vec.append(uni_x)
    array = np.array(vec)
    
    #make distribution mean 0 per hw directions
    npvec = array - (np.mean(array))

    #normalizes vector by dividing by vector length/magnitude
    mag = math.sqrt(sum(npvec ** 2))
    out = npvec / mag 
    return out
        
```

```python
def many_dots(pairs):
    '''
    Computes dot product of randomly generated unit vectors from a uniform distribution at different dimensions.
    pairs: number of dot products for each dimension
    
    returns:
        a list of dot products
    '''
    dim = [10,20,50,100,250,500,1000,2000]
    out = []
    for i in dim:
        for j in range(pairs):
            vec1 = ran_vector(i)
            vec2 = ran_vector(i)
            prod = np.dot(vec1,vec2)
            out.append(prod)
    return out
        
```

```python
dot_products = many_dots(50000)
```

```python
def ran_coord(n):
    '''
    Generates n number of random coordinates from -1 to 1 from uniform distribution.
    Outputs a dataframe with X and Y columns.
    n: number of random numbers
    '''
    X = []
    Y = []
    for i in range(n):
        X1 = rnd.uniform(-1,1)
        Y1 = rnd.uniform(-1,1)
        X.append(X1)
        Y.append(Y1)
    d = {"X": X, "Y":Y}
    df = pd.DataFrame(data=d)
    return df
        
        
```

```python
def in_circle(n):
    '''
    Computes an estimate of 4/pi using the Monte Carlo method
    
    n: numbers of randomly generated points from uniform distribution to classify as in or out of the circle
    
    returns:
        Two dataframes. 1st with the coordinates that fall inside circle. 2nd with the coordinates that fall outside.

    '''

    X = []
    Y = []
    for i in range(n):
        X1 = rnd.uniform(-1,1)
        Y1 = rnd.uniform(-1,1)
        X.append(X1)
        Y.append(Y1)
    d = {"X": X, "Y":Y}
    df = pd.DataFrame(data=d)
      
    true_x =[]
    true_y =[]
    
    false_x = []
    false_y = []
    
    for i in range(len(df)):
        if ((df["X"][i] - 0) * (df["X"][i]  - 0) +
            (df["Y"][i] - 0) * (df["Y"][i]  - 0) <= 1):
                true_x.append(df["X"][i])
                true_y.append(df["Y"][i])   
        else:
            false_x.append(df["X"][i])
            false_y.append(df["Y"][i])
    true = pd.DataFrame({"X": true_x, "Y": true_y})
    false = pd.DataFrame({"X": false_x, "Y": false_y})
    return (true, false)
                   
                
 
```
