```python
def encode_name(data):
    '''
    Encodes 1,2,3 if name starts with consonant,consonant (1),
    consonant,vowel (2), contains a number (3)
    
    '''
    
    kl_pattern = r'^[bcdfghjklmnpqrstvwxyz]{2}\w*'
    kl_mask = data["Name"].str.findall(kl_pattern,re.IGNORECASE)
    kl_indices = np.where(kl_mask)[0]
    data["Encoded Name"][kl_indices] = 0
    
    ro_pattern = r'\b[b-df-hj-np-tv-z][aeiou]\w*\b'
    ro_mask = data["Name"].str.findall(ro_pattern,re.IGNORECASE)
    ro_indices = np.where(ro_mask)[0]
    data["Encoded Name"][ro_indices] = 1
    
    an_pattern = r'\d+'
    an_mask = data["Name"].str.findall(an_pattern,re.IGNORECASE)
    an_indices = np.where(an_mask)[0]
    data["Encoded Name"][an_indices] = 2
    
    return data
    
```

```python
def planet_codes(n):
    '''
    n: number of zeros after ship code to match number of features
    '''
    klingon = np.zeros((n))
    klingon[0] = 1
    
    romo = np.zeros((n))
    romo[1] = 1
    
    ant = np.zeros((n))
    ant[2] = 1
    
    fed = np.zeros((n))
    fed[3] = 1
   
    x = np.concatenate((np.stack((klingon,) * 5, axis=0), np.stack((romo,) * 5, axis=0), np.stack((ant,) * 5, axis=0),np.stack((fed,) * 5, axis=0) ))
    return x
    
    
```

```python
def mean_squared_error(pred, gs):
    pred = np.array(pred)
    gs = np.array(gs)
    squared_errors = np.sum((pred - gs) ** 2, axis=1)
    mse = np.mean(squared_errors)
    return mse
```

```python
def Widrow_Hoff(k, g, g_prime, f, dim):
    length_f = np.linalg.norm(f)
    difference_vector = np.subtract(g, g_prime)
    deltaA = k * (np.outer(difference_vector, np.transpose(f)))
    return deltaA, difference_vector

```

```python
def LMS_ship (data,ship_code,threshold,k,ksub):
    '''
    Uses the Least Means Squares Algorithm to correct matrix A until
    change in mean square error between iterations is below a threshold.
    
    info: characteristics of ships
    ship_code: encoded planet for ships
    threshold: What the change in mean square error is for simulation to stop
    
    '''
    
    f_set = df.to_numpy()
    g_set = ship_code
    pairs = len(ship_code)
    dim = len(f_set[0])
    
    #intialize A with zeros
    #A = np.zeros((pairs,dim))
    
    pred_list = []
    gs = []
    norms = [0]
    count = 1 #keeps track of how many time while loop occurs, has to start w 1 bc of K
    sq_errors = [1] # keeps track of mean standard error
    all_mse = []
    Errors=[]
    converged = False
    m = len(f_set[0])
    n = len(g_set[0])
    
    A_i = [ np.outer(f_set[i],np.transpose (g_set[i])) for i in range(pairs) ]
    A = np.zeros((m,n))
    for i in range(pairs):
        A = np.add(A, A_i[i])
    
    
    while converged == False:# number of iterations of this loop is number of trials
        randomList = [] #keeps track of which pairs used in one trial
        for i in range(pairs-len(randomList)):
            #make sure a random, but not sequential pair is picked
            #also ensures entire group is being picked from, without replacement
            
            r = rnd.randint(1,pairs-1)
            if r not in randomList: 
                randomList.append(r)
                
                #Pick a random pair t
                f = f_set[r]
                g = g_set[r]
                g_prime = A.dot(f.T)
                
                pred_list.append(g_prime)
                
                #keep tracks of g vectors used
                gs.append(g)

                
                #set K to constant / # of learning trials
                #k = ((1/np.dot(f,f)) - (1/1000)) / trials
                
                #updating of A and predictions
                deltaA, error = Widrow_Hoff(k, g, g_prime, f, dim)
                A = np.add(A, deltaA)
                error_length = np.linalg.norm(error)
                Errors.append(error_length)
                #print("Current Error:", error_length)

                k =  abs(k - ksub)
            
            #continues for loop if pair is already picked
            else:
                continue
            mse = mean_squared_error(pred_list, gs)
            all_mse.append(mse)
            print("Current MSE:", mse)
        #if (abs(error_length) < threshold):
        if (abs(mse) < threshold):
            print("MSE: ", mse)
            #print("Error: ", abs(error_length))
            print("Error Threshold: ", threshold)
            print("Number of trials: ", count)
            
            return A
            converged = True
        
    count = count + 1;
    
```

## Write Up

### f vectors
The quantative data(Warp Drive, Hailing Transponder, and Ratio of axis) stayed the same in the feature vectors. To represent the name of the ships, I classified them into if they started with a (consonant, consonant), (consonant,vowel), or (if it contained a number / starts with a capital A or E). In this order, these categories were encoded into 0,1, or 2. The Surface Reflect color was encoded into 0 = Black,Dark Gray = 2, Dark Green = 3,Light Blue = 4, Light Gray = 5, Orange = 6, Pink = 7, White = 8, Yellow = 9. The order of the feature vector followed the order of the data frame (Name, Warp Drive, Hailing Transponder, Surface Reflect, and Ratio of axis). I decided to use all vectors since from my own judgement, it seemed like each planet had something unique in one of the catagories, like Romulan's always having a 3 digit Hailing Transponder Freq. according to the train data. 

### g vectors

I coded the different planets of each ship using one hot encoding.
The number of elements in the vector corresponded with the number number of elements in the feature vector. Klingon was [1,0,0,0,0], Romulan was [0,1,0,0,0], Antarean [0,0,1,0,0], and Federation was [0,0,0,1,0]. To decode the output of the system as guesses for each planet, I either found where the value was 1 and a little more, or found the first positive value that occured and counted that as a 1. Everything else was considered a 0.
